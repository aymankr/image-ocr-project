## Ne pas confondre : résolution et définition

* la définition est la taille, en pixel, d’une image sur un écran, un site web
* la résolution, en dpi, (combinée à la définition) caractérise la taille d’une image lors d’une impression.

```python
original_image = data.chelsea()

def show_images(before, after, op):
    fig, axes = plt.subplots(nrows=1, ncols=2)
    ax = axes.ravel()
    ax[0].imshow(before, cmap='gray')
    ax[0].set_title("Original image")

    ax[1].imshow(after, cmap='gray')
    ax[1].set_title(op + " image")
    if op == "Rescaled":
        ax[0].set_xlim(0, 400)
        ax[0].set_ylim(300, 0)
    else:        
        ax[0].axis('off')
        ax[1].axis('off')
    plt.tight_layout()
        
# rescale image to 25% of the initial size
image_rescaled = rescale(original_image, 0.499999)

show_images(original_image, image_rescaled, "Rescaled")
#from https://www.kaggle.com/tomahim/image-manipulation-augmentation-with-skimage
```

## histogramme d'une image 

L’histogramme d’une image h(x) est la fonction qui associe à une valeur d’intensité x le nombre de pixels dans l’image ayant cette valeur.

## Luminance ou brillance d’une image
* La luminance (ou brillance) est définie comme la moyenne de tous les pixels de l’image.
* Il suffit donc de décaler l’histogramme en modifiant les NdG de l’image : I (x , y ) = I (x , y ) + b
* Dans les deux images suivantes, seule la luminance est différente

```python
import numpy as np
image = data.page()

#Par defaut UINT
image=image.astype(int)
value=50
imageDarker = image-value
imageBrighter = image+value

#rester entre 0 et 255
imageDarker[imageDarker < 0] = 0
imageBrighter[imageBrighter > 255] = 255

np.clip(imageDarker, 0, 255,imageDarker)
np.clip(imageBrighter, 0, 255,imageBrighter)

fig, ax = plt.subplots(3, 2, figsize=(10, 10))

ax[0,0].imshow(image, cmap=plt.cm.gray)
ax[0,0].set_title('Original')

ax[0,1].hist(image.ravel(), bins=256)
ax[0,1].set_title('Histogram')

ax[1,0].imshow(imageBrighter, cmap=plt.cm.gray)
ax[1,0].set_title('Brighter')

ax[1,1].hist(imageBrighter.ravel(), bins=256)
ax[1,1].set_title('Histogram')


ax[2,0].imshow(imageDarker, cmap=plt.cm.gray)
ax[2,0].set_title('Darker')

ax[2,1].hist(imageDarker.ravel(), bins=256)
ax[2,1].set_title('Histogram')


for a in ax[:, 0]:
    a.axis('off')
plt.show()
```

## Changement de contraste
On préfèrera changer le contraste plutôt que la luminance
* Transformation linéaire
* Transformation non-linéaire

### Formule du contraste

Le contraste peut être défini de plusieurs façons :
* Ecart-type des niveaux de gris (N=nombre de lignes, M=nombre de colonnes)
$$\sqrt{\frac{1}{NM}\displaystyle\sum_{x=1}^N \sum_{y=1}^M (I(x,y)-Moy)^2}$$
* Variation entre niveaux de gris max et min
$$\frac{max[f(x,y)]-min[f(x,y)]}{max[f(x,y)]+min[f(x,y)]}$$

## Egalisation d'histogramme (automatique)



* Calcul de l’histogramme $h(i)$ avec $i \in [0, 255]$
* Normalisation de l’histogramme $h_n(i) = h(i)/N$ avec $i \in [0, 255]$
* Densité de probabilité cumulative $C(i) = \sum_{j=1}^{i} {h_n(j)}$ avec $i \in [0, 255]$
* Transformation des niveaux de gris de l’image $f′(x,y)= C(f(x,y))∗255$

Si f et g sont deux images, on peut définir l’addition R pixel à pixel de ces deux images par :
$$R(x,y)=Min(f(x,y)+g(x,y); 255)$$

L’addition d’images peut permettre
* De diminuer le bruit d’une vue dans une série d’images
* D’augmenter la luminance en additionnant une image avec elle-même

<img style="text-align:left" src="images_slides/toConvert/transfo29.png" width=80%>


On peut définir la soustraction S pixel à pixel de deux images f et g par :
$$ S(x,y)=Max(f(x,y)−g(x,y);0)$$

La soustraction d’images peut permettre 
* Détection de défauts
* Détection de mouvements


<img style="text-align:left" src="images_slides/toConvert/transfo30.png" width=80%>



La multiplication S d’une image f par un ratio (facteur) peut se définir par :
$$ S(x,y)=Max(f(x,y)∗ratio;255) $$
La multiplication d’images peut permettre d’améliorer le contraste ou la luminosité
